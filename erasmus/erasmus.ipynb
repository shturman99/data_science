```python
X = data[["Countries","Departments","Exam score"]]
Y = data[["Grant"]]
```

```python
X
```

```python
from sklearn import preprocessing
labelencouder = preprocessing.LabelEncoder()
```

```python
X["Countries"]= labelencouder.fit_transform(data["Countries"])
```

```python
labelencouder.classes_[0]
```

```python
X["Departments"]= labelencouder.fit_transform(data["Departments"])
```

```python
labelencouder.classes_
```

```python
from sklearn.model_selection import train_test_split

X_test,X_trian, y_test, y_trian = train_test_split(X,Y,train_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()

log_model.fit(X_trian,y_trian.values.ravel())
```

```python
from sklearn.metrics import accuracy_score
```

```python
log_model_prediciton = log_model.predict(X_test)
```

```python
accuracy_score(y_test,log_model_prediciton)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.svm import SVC

svm = SVC()

par_grid_svc = {"C": [0.01,0.1,0.5],"kernel":["linear","rbf","poly"]}
```

```python
gridsvc = GridSearchCV(svm,par_grid_svc)
```

```python
gridsvc.fit(X_trian,y_trian.values.ravel())
```

```python
gridsvc.best_params_
```

```python
gridsvc.best_estimator_.predict
```

```python
svc_prediction = gridsvc.predict(X_test)
```

```python
svc_prediction
```

```python
accuracy_score(y_test,svc_prediction)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier()
n_estimators = [32,64,128,256]
max_features = [2,3,4]
par_grid_rfc ={"n_estimators":n_estimators,"max_features":max_features}
```

```python
gridrfc = GridSearchCV(rfc_model,par_grid_rfc)
```

```python
gridrfc.fit(X_trian,y_trian.values.ravel())
```

```python
gridrfc.best_params_
```

```python
prediction_rfc = gridrfc.predict(X_test)
```

```python
accuracy_score(y_test,prediction_rfc)
```
